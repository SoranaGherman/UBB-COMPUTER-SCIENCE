```python
import numpy as np 
import matplotlib.pyplot as plt 
import math
 
angle = np.linspace( 0 , 2 * np.pi , 100 ) 
 
x = np.cos(angle) 
y = np.sin(angle) 
r = 1
 
figure, axes = plt.subplots(1) 
 
axes.plot(x, y) 
axes.set_aspect(1) 

x = np.array([0,120,240,0])
y = np.array([0,120,240,0])
x = np.cos(x * np.pi / 180)
y = np.sin(y * np.pi / 180)

plt.plot(x, y, color='red')


fav_cases = 0
total_cases = 100

for i in range(100):
    arr = np.random.random(2) 
    alpha1 = arr[0] * (2 * np.pi)
    x2, y2 = r * np.cos(alpha1), r * np.sin(alpha1)
    plt.plot(x2, y2, color = 'red', marker = '.')
    
    alpha2 = arr[1] * (2 * np.pi)
    x3, y3 = r * np.cos(alpha2), r * np.sin(alpha2)
    
    plt.plot(x3, y3, color = 'green', marker = '.')
    
    x4 = [x2, x3]
    y4 = [y2, y3]
    
    plt.plot(x4, y4)
    
    dist = math.sqrt((x2 - x3) ** 2 + (y2 - y3) ** 2)
    triangle_side = math.sqrt((x[0] - x[1]) ** 2 + (y[0] - y[1]) ** 2)
    
    if dist > triangle_side:
        fav_cases += 1
        plt.plot(x4, y4, color = 'green')
    else:
        plt.plot(x4, y4, color = 'blue')

prob = fav_cases / total_cases
print(prob)

plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt 
import math
 
angle = np.linspace( 0 , 2 * np.pi , 100 ) 
 
x = np.cos(angle) 
y = np.sin(angle) 
r = 1
 
figure, axes = plt.subplots(1) 
 
axes.plot(x, y) 
axes.set_aspect(1) 

x = np.array([0,120,240,0])
y = np.array([0,120,240,0])
x = np.cos(x * np.pi / 180)
y = np.sin(y * np.pi / 180)

plt.plot(x, y, color='red')

fav_cases = 0
total_cases = 100


for i in range(100):
    # random radius
    arr = np.random.random(1) 
    alpha = arr[0] * (2 * np.pi)
    xA, yA = r * np.cos(alpha), r * np.sin(alpha)
    plt.plot([0, xA], [0, yA], color = 'black')
    
    r_middlex, r_middley = r / 2 * np.cos(alpha), r/2 * np.sin(alpha)  # the coord of the middle of the radius

    # random point on radius
    arr = np.random.random(1) 
    alpha2 = arr[0]
    random_point_on_radius = r * alpha2
    xB, yB = random_point_on_radius * np.cos(alpha), random_point_on_radius * np.sin(alpha)
    plt.plot(xB, yB, color = 'green', marker = '.')

    OB = math.sqrt(xB ** 2 + yB ** 2)
    BC = math.sqrt(1 - OB ** 2)

    xC = xB + yA * BC
    yC = yB - xA * BC

    xD = xB - yA * BC
    yD = yB + xA * BC


    plt.plot([xC, xD], [yC, yD], color = 'green')
    
    if math.sqrt(r_middlex ** 2 + r_middley ** 2) < math.sqrt(xB ** 2 + yB ** 2):
        plt.plot([xC, xD], [yC, yD], color = 'green')
    else:
        plt.plot([xC, xD], [yC, yD], color = 'orange')
        fav_cases += 1

prob = fav_cases / total_cases
print(prob)

plt.show()
    
    
```

```python
import numpy as np 
import matplotlib.pyplot as plt 
import math, random
 
angle = np.linspace( 0 , 2 * np.pi , 100 ) 
 
x = np.cos(angle) 
y = np.sin(angle) 
r = 1
 
figure, axes = plt.subplots(1) 
 
axes.plot(x, y) 
axes.set_aspect(1) 

x = np.array([0,120,240,0])
y = np.array([0,120,240,0])
x = np.cos(x * np.pi / 180)
y = np.sin(y * np.pi / 180)

plt.plot(x, y, color='red')


# here we are going to draw the smaller circle

circle = plt.Circle((0, 0), r/2, color='blue', fill=False)
axes.add_artist(circle)

fav_cases = 0
total_cases = 100

for i in range(100):
    # a new point within the circle
    ok = 0
    while ok == 0:
        xB =  random.random() * 2 - 1
        yB =  random.random() * 2 - 1
        
        if math.sqrt(xB ** 2 + yB ** 2) <= r:
            ok = 1

    
    plt.plot(xB, yB, color='black', marker='.')
    
    alpha = np.arctan(yB/xB)
    xA, yA = r * np.cos(alpha), r * np.sin(alpha)
    
    OB = math.sqrt(xB ** 2 + yB ** 2)
    BC = math.sqrt(1 - OB ** 2)

    xC = xB + yA * BC
    yC = yB - xA * BC

    xD = xB - yA * BC
    yD = yB + xA * BC
    
    if math.sqrt(xB ** 2 + yB ** 2) <= r / 2:
        plt.plot([xC, xD], [yC, yD], color='green', marker='.')
        fav_cases += 1
    else:
        plt.plot([xC, xD], [yC, yD], color='blue', marker='.')
    

prob = fav_cases / total_cases
print(prob)

plt.show()
    
    
```
